```python
from diffusers import DiffusionPipeline
import torch

pipeline = DiffusionPipeline.from_pretrained("nicky007/stable-diffusion-logo-fine-tuned")

pipeline.to("cuda")
```

```python
prompt = "A logo for a coding hackathon, featuring a stylized laptop or code brackets signifying coding and digital culture"
```

```python
seed = 2
image = pipeline(prompt=prompt, num_inference_steps=50, guidance_scale=7.5, seed=seed).images[0]
image
```
